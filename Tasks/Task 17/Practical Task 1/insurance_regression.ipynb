```python
#Import necessary libraries:
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Load the dataset
insurance_df = pd.read_csv('insurance.csv')
```

```python
# Display the first few rows of the DataFrame
print(insurance_df.head())

# Check the column names and data types
print(insurance_df.info())
```

```python
# Plot a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(insurance_df['age'], insurance_df['charges'], alpha=0.5)
plt.title('Scatter Plot of Age vs Charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.grid(True)
plt.show()
```

```python
# Extract features (X) and target variable (y)
X = insurance_df[['age']]
y = insurance_df['charges']

# Initialise and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions
predictions = model.predict(X)
```

```python
# Plot scatter plot with best-fit line
plt.figure(figsize=(10, 6))
plt.scatter(insurance_df['age'], insurance_df['charges'], alpha=0.5)
plt.plot(X, predictions, color='red', linewidth=2)
plt.title('Scatter Plot of Age vs Charges with Best-Fit Line')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.grid(True)
plt.show()
```
