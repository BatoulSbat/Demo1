```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data
df = pd.read_csv('Cars93.csv')
```

### 1. Boxplot

```python
# Boxplot for revs per mile
selected_manufacturers = ['Audi', 'Hyundai', 'Suzuki', 'Toyota']
df_selected = df[df['Manufacturer'].isin(selected_manufacturers)]
plt.figure(figsize=(10, 6))
plt.boxplot([df_selected[df_selected['Manufacturer'] == m]['Rev.per.mile'] for m in selected_manufacturers],
            labels=selected_manufacturers)
plt.title('Revs per mile for selected car manufacturers')
plt.ylabel('Revs per mile')
plt.xlabel('Manufacturer')
plt.show()
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?

**Boxplot answer:**
Based on the box plot graph, it appears that Toyota might have the car with the highest revs per mile. This is because Toyota’s box plot displays significant variability with a much larger interquartile range (IQR) and outliers indicated by dots above the upper whisker.

### 2. Histogram

```python
# Histogram of MPG
plt.figure(figsize=(10, 6))
plt.hist(df['MPG.city'], bins=20, alpha=0.5, label='MPG City')
plt.hist(df['MPG.highway'], bins=20, alpha=0.5, label='MPG Highway')
plt.title('MPG Distribution in City and Highway')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**
Based on the bar graph, it is generally more fuel efficient to drive on the highway than in the city. The graph shows MPG (miles per gallon) distribution in city and highway driving conditions. The blue bars represent city driving conditions, with most frequencies falling between 10 and 20 MPG. The orange bars represent highway driving conditions, with most frequencies falling between 20 and 30 MPG. Therefore, higher MPG values, indicating better fuel efficiency, are more frequent during highway driving compared to city driving. However, actual fuel efficiency can vary depending on factors like driving style, vehicle condition, and traffic conditions.

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
plt.figure(figsize=(10, 6))
plt.plot(df['Wheelbase'], df['Turn.circle'], marker='o', linestyle='-')
plt.title('Relationship between Wheelbase and Turning Circle')
plt.xlabel('Wheelbase')
plt.ylabel('Turning Circle')
plt.grid(True)
plt.show()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?

**Lineplot answer:**  
The scatter plot shows the relationship between the wheelbase of a vehicle and its turning circle. The wheelbase is the distance between the centers of the front and rear wheels, while the turning circle is the smallest circular turn that the vehicle can make.

From the graph, it appears that there is a positive correlation between wheelbase and turning circle. This means that as the wheelbase gets larger, the turning circle also increases. In practical terms, a vehicle with a larger wheelbase would require more space to complete a turn compared to a vehicle with a smaller wheelbase. This is because a longer wheelbase can lead to a larger turning radius, resulting in a larger turning circle. However, this is a general trend and actual results can vary depending on other factors such as the design of the vehicle and its steering mechanism.

### 4. Barplot

```python
# Barplot for mean horsepower
mean_horsepower = df.groupby('Type')['Horsepower'].mean()
plt.figure(figsize=(10, 6))
mean_horsepower.plot(kind='bar')
plt.title('Mean Horsepower for each Car Type')
plt.xlabel('Car Type')
plt.ylabel('Mean Horsepower')
plt.xticks(rotation=45)
plt.show()
```

**Barplot question:**  
Does a larger car mean more horsepower?

**Barplot answer:**
Based on the bar chart, it appears that a larger car does not necessarily mean more horsepower. The chart shows different car types and their mean horsepower. While some larger types like “Large” and “SUV” have high mean horsepower, the “Midsize” category has less than “Sporty,” which may not be considered a large car type.

Here’s a breakdown of the mean horsepower for each car type based on the chart:

Compact: Lower mean horsepower
Large: Highest mean horsepower
Midsize: Significantly lower mean horsepower than both “Large” and “SUV”
Small: Lower mean horsepower
Sporty: Higher mean horsepower, even though it may not be considered a large car type
SUV: High mean horsepower, close to “Large”

So, while larger cars like “Large” and “SUV” categories tend to have higher horsepower, it’s not a strict rule. For instance, “Sporty” cars, which are typically not large, also have high horsepower. Therefore, the size of a car does not always directly correlate with its horsepower. Other factors such as the car’s engine type, design, and technology also play a significant role in determining its horsepower.
