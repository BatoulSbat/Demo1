```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Titanic dataset
titanic_df = pd.read_csv('titanic.csv')
```

```python
# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(titanic_df.head())
```

```python
# Check for missing values
print("\nMissing values:")
print(titanic_df.isnull().sum())
```

```python
# Imputation for Age
median_age = titanic_df['Age'].median()
titanic_df['Age'].fillna(median_age, inplace=True)

# Imputation for Embarked
most_common_embarked = titanic_df['Embarked'].mode()[0]
titanic_df['Embarked'].fillna(most_common_embarked, inplace=True)
```

```python
# Check for missing values after handling
print("\nMissing values after handling:")
print(titanic_df.isnull().sum())
```

```python
# What is the most important factor in determining survival of the Titanic incident?
# Analyse survival based on different factors
plt.figure(figsize=(10, 8))
sns.countplot(x='Survived', hue='Sex', data=titanic_df)
plt.title('Survival by Gender')
plt.show()

plt.figure(figsize=(10, 8))
sns.countplot(x='Survived', hue='Pclass', data=titanic_df)
plt.title('Survival by Passenger Class')
plt.show()
```

```python
# In the movie, the upper-class passengers were given preference on lifeboats. Does this show in the data?
# Analyse survival rates by passenger class
plt.figure(figsize=(10, 8))
sns.countplot(x='Pclass', hue='Survived', data=titanic_df)
plt.title('Survival by Passenger Class')
plt.show()
```

```python
# "Women and children first". Was this the case?
# Analyse survival rates by gender and age
plt.figure(figsize=(10, 8))
sns.countplot(x='Survived', hue='Sex', data=titanic_df)
plt.title('Survival by Gender')
plt.show()

plt.figure(figsize=(10, 8))
sns.histplot(x='Age', hue='Survived', data=titanic_df, kde=True)
plt.title('Survival by Age')
plt.show()
```

```python
# Add one other observation that you have noted in the dataset.
# analyse survival rates by fare and cabin location
plt.figure(figsize=(10, 8))
sns.histplot(x='Fare', hue='Survived', data=titanic_df, kde=True)
plt.title('Survival by Fare')
plt.show()

plt.figure(figsize=(10, 8))
sns.countplot(x='Cabin', hue='Survived', data=titanic_df)
plt.title('Survival by Cabin')
plt.show()
```
