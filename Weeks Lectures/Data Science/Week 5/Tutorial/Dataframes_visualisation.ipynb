```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#check availabile datasets
sns.get_dataset_names()

#three ways of reading in the data

#loading the dataset from the seaborn library (online dataset)
#buit in titanic dataset
df = sns.load_dataset("titanic")

# #fetch data from github
# url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv'
# df = pd.read_csv(url)

# #read csv file from local diractory, for this way we need to download the file. 
# filename = 'titanic'
# df = pd.pd.read_csv(filename)
```

```python
#drop null values
new_df = df.dropna()
df.info()
```

```python
#replace the null values with mean, median or mode
x = df['age'].mean()
df['age'].fillna(x, inplace = True) 
#inplace = True is for overwriting the dataframe, replaces the null value with the mean 
#and puts the value(mean) instead of the null one 
```

```python
df.info()
```

Visualisation

```python
sns.histplot(df['age'], bins = 30, color = 'g', kde = True) #kernal density plots (the line)
plt.savefig('hist.png')
```

```python
#removing rows
for x in df.index:
    if df.loc[x, 'age'] > 50:
        df.drop(x, inplace = True)
    

sns.histplot(df['age'], bins = 30, color = 'yellow', kde = True)
```

```python
sns.kdeplot(df['age'], fill = True)
```
